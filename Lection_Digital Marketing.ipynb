# Стандартные задачи в области перформанс-маркетинга

# Задача 1

К нам пришел клиент, который хочет получить А кликов за свои Х рублей. 
За первую половину срока рекламной кампании он получил В (В<А) кликов и потратил Z (Z<X) рублей.

**Вопрос:** Какой цены клика в дальнейшем нужно придерживаться, чтобы клиент получил все запланированные клики? 
**Решение:** (X - Z) / (A - B)


# Задача 2

К нам пришел клиент, который планирует запустить кампанию в определенный день на определенный срок Y с бюджетом Х. 
Но через какое-то время клиент решил изменить (увеличиьть/уменьшить) срок на Z дней. 

**Вопрос:** Сколько денег должно тратиться в день, чтобы кампания равномерно крутилась весь срок? 

**Решение:** X / (Y + Z)

# Задача 3

Рекламный бюджет: 40 000 рублей. Уровень конверсии сайта в клиентов 1% (<=> чтобы получить 1 клиента, надо завлечь 100 посетителей). Стоимость привлеченного клиента должна быть не более 300 рублей. 

**Вопрос:** Сколько переходов на сайт нужно получить клиенту в рамках бюджета? 

**Решение:** 40 000 / 300 = X * 0.01

**MMM - Media Mix Modelling**

Статистическое моделирование эффективности инвестиций в рекламу на основе данных прошлых периодов

**Продажи = Базовые продажи + Реклама + Другие факторы + Погрешность**

базовые продажи = что и так купили бы (при отключении затрат на маркетинг вообще)

реклама: paid + owned + earned, где

paid - традиционная (ТВ, радио, ООН, печатная, спонсорная и т.д.) и digital (социальные сети, дисплей, видео реклама, контекстная реклама)
owned - веб-сайт, группы в соц. сетях, e-mail и т.д.
earned - PR, органический поиск/SEO, сарафанное радио и т.д.

другие факторы = на которые вы можете повлиять (продукты/ промо/ дистрибуция/ сервис/ цена) и outside advertiser control (тренды/ погода/ активность конкурентов/ экономика/сезонность)

    - аудитории по интересам (любители кино, зимние виды спорта, Light TV viewers, ...)
