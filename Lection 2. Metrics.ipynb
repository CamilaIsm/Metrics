Прокси-метрика — это косвенная мера целевой метрики, с которой она сильно коррелирует. По изменению в прокси-метрике мы как минимум должны понять направление изменения целевой метрики. 

Recency (метрика) - давность последнего контакта с пользователем

**GMV (gross merchandise volume)**  - общий объем оборота товаров - это совокупность продаж в денежном выражении, осуществленных на электронной торговой площадке в течение определенного периода времени. Это реальный валовый показатель объема покупок, совершенных на электронной торговой площадке. Это также полезный показатель объема потенциала торговой площадки и может быть использован в качестве показателя потенциальной прибыли при пересчете в годовое исчисление.


Важно понимать, что выручка — это производная от всего оборота средств в компании (GMV, gross merchangise volume). Компании оцениваются не по выручке. Маржинальность в разных бизнесах может принимать различные значения, но вне зависимости от выручки компания продолжает расти.


**AOV (Average Order Value) и RPR (Repeat Purchase Ratio)**
AOV — это средний размер одной транзакции, общий AOV по платформе считается просто — делите весь GMV на количество транзакций. RPR же показывает, какая часть покупателей совершает повторные покупки. По факту для маркетплейсов это некий аналог retention. Если вы видите, что большинство покупателей делает только одну транзакцию (хотя по логике они должны быть регулярными), значит, скорее всего, покупатели либо не получают необходимую ценность, либо после первого удачного взаимодействия просто “отваливаются” и начинают сотрудничать с поставщиком вне платформы, чтобы не платить комиссию (а это тоже губит множество маркетплейсов).
AOV и RPR позволяют понять поведенческие паттерны потребителей, а вместе с take rate (комиссией и другими источниками дохода маркетплейса) — становятся базисом для расчета LTV.

- AOV = GMV / кол-во транзакций 
- RPR = кол-во пользователей, совершивших более 1 покупки / кол-во пользователей

# CAC = сумма потраченного рекламного бюджета на привлечение новых клиентов / количество привлеченных клиентов

**Retention** - удержание пользователей. Метрика показывает, сколько пользователей продолжает использовать продукт через определенный период времени, и напрямую влияет на пожизненную ценность клиента (LTV).  

# Retention rate  = количество активных пользователей на конец периода / количество активных пользователей на начало периода * 100%

Пример: пусть из 1000 юзеров, которые привлекла компания, 500 были активны в первый день, а на седьмой день - 200. Коэффициент удержания седьмого дня = 200/500 * 100% = 40%.

**CMGR / CAGR (Compounded monthly / Annual growth rate)** - среднемесячный / среднегодовые темпы роста чего-либо (чаще всего выручки)

**ARR (annual recurring revenue) / MRR** - годовая или месячная возобновляемая выручка - одна из составляющих выручки, которая возобновляется регулярно 

**Users (DAU, MAU)** - посещаемость, ежедневные или ежемесячные активные пользователи

**Churn** - отток - доля пользователей, отказавшихся от использования продукта

**Burn Rate** - ежемесячная потеря денег компанией, усредненный убыток. Если количество денег в банке поделить на burn rate получится количество месяцев, которые компания сможет остаться в бизнесе

# Метрики роста

Если прибыль не является главным показателем, то как оценить положение дел на
метриках?

Этот фреймворк хорош на раннем этапе, когда вы тестите какую-то гипотезу или у вас есть небольшой работающий бизнес.

-  **New Users** — количество пользователей, которые начали пользоваться вашим продуктом. Все зависит от затрат на маркетинг, но для нового продукта их количество должно увеличиваться. Если количество новых юзеров неэластично по затратам на маркетинг, то у вас проблемы с ЦА или с продуктом.
- **С1** — конверсия в первое целевое действие. Если пользователь совершил покупку, мы понимаем, решил ли он свою задачу и нашел ли то, что искал
- **CN** — конверсия в очередное действие. Сопоставив, как С10 (С50, С1000) коррелирует с конверсией регулярного использования, мы можем приблизительно определить, после какого количества покупок пользователь становится  нашим 
- **Frequency** — частота использования. Понятно, что не для всех бизнесов такая метрика будет работать — например, в случае Авито.Недвижимость человек не так часто покупает квартиру, чтобы регулярно пользоваться приложением.
- **Churn Rate** — количество потерянных пользователей. Почему смотрим Churn, a не Retention? Эта метрика важна не количественно, а качественно. При отслеживании этой метрики можно увидеть, на каком этапе пользователи  отваливаются , и это дает возможность изучить возможные причины их ухода. Это особенно важно для новых продуктов.


# Метрики здоровья прокукта / сервиса

- CAC - цена привлеченного клиента. Эту метрику можно разбить на две: на цену единожды совершившего целевое действие пользователя и на конвертацию в постоянного пользователя. 

**САС** = сумма потраченного рекламного бюджета на привлечение клиентов / количество привлеченных клиентов

- Retention Rate / Churn Rate - коэффициент удержания / потери клиента. Показывает, сколько пользователей продолжает использовать продукт через определенный срок, и напрямую влияет на пожизненную ценность клиента (LTV).

**Retention rate** = количество активных пользователей на конец периода / количество активных пользователей на начало периода * 100%

- LTV - пожизненная ценность клиента. Полезен для сегментации аудитории и персонификации подходов. Удобно сравнивать
с CAC — если где-то у нас LTV меньше CAC, значит, мы что-то делаем не так.

- DAU / MAU — количество уникальных пользователей за день / месяц. В некотором смысле эта метрика  зашита  в ретеншн. Эта метрика показывает вовлеченность пользователя, позволяет исследовать частоту использования. Сравнение отношений DAU/MAU часто используется в gamedev и частотных приложениях, например, для дейтинга.

- Net promoter score / CSAT - метрика, позволяющая оценивать удовлетворенность клиента.

# Метрики доходов

- Revenue — выручка, полученная на продаже продукта. Однако выручка может быть отрицательной. Если вы — растущий бизнес, то фокусируетесь на росте GMV.
- GMV, gross merchangise volume — валовый объем торговых операций, оборот денежных средств в компании. В этой же метрике измеряется доля рынка.
- ARPU, average revenue per user — средняя прибыль от одного пользователя за определенный срок.
- ARPPU, average revenue per paying user — средняя прибыль от одного платящего пользователя за определенный срок

Основные метрики про доходы — это все же выручка и оборот. ARPU и ARPPU — более узкоспециализированные метрики. Они используются, например, в сервисах, в которых неплатящие пользователи могут конвертироваться в платящих (например, бесплатная подписка и премиум).
