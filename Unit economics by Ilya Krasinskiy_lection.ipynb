# Метрики

**Лекция о юнит-экономике Ильи Красинского (https://youtu.be/h8VWl0GFW3Y)**

3 главных инсайта из лекции
1. Не всегда «человек рациональный» должен быть предпосылкой перед началом анализа данных о поведении пользователя (на сайте/ в приложении). Человек зачастую принимает решения, основываясь на эмоциях, и это нельзя не принимать во внимание. Также очень важно уметь абстрагироваться и не оценивать пользователей по себе, иметь четкое представление кто наша целевая аудитория
2. В алгоритме поиска точек роста начинать необходимо не с идеи, а с диагностики
3. Информативность показателя «конверсия во 2-ю покупку» (С2). Этот показатель позволяет нам отсечь случайных пользователей, купивших продукт 1 раз и больше не заходивших на сайт. Также интересным показалось то, что Илья, при расчете конверсии в покупку (%), домножал ее на конверсию от регистрации на сайте(%), и уже на основании этого показателя делал последующие выводы.  


  В начале лекции Илья сравнивает данные по двум приложениям - Bookmate и Литрес. И приходит к выводу , что если продукт начал терять позиции на рынке , необходимо обратить внимание на дизайн (сайта/приложения) и подумать, насколько удобен и красив интерфейс с точки зрения пользователя. Возможно дело не в какой-то отдельно взятой фиче, а в дизайне мобильного приложения (сайта) в целом. 
  

Алгоритм поиска точек роста (идея - реализация - измерение - анализ):
1. Диагностика: найти метрику, изменения которой дают наибольший эффект
2. Выдвинуть гипотезы
3. Реализовать гипотезы
4. Посчитать эффект, сделать выводы
