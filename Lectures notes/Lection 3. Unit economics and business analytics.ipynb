# Метрики

**Лекция 3. Юнит экономика и бизнес аналитика**

# Базовая модель монетизации

Переменные затраты - это те, которые способствуют увеличению продаж и привлечению клиентов, они растут пропорционально выручке.

Постоянные затраты необходимы для поддержания бизнес-процессов (персонал, офис). Они не имеют прямой зависимости от выручки, количества транзакций и среднего чека. При масштабировании бизнеса такие затраты растут поэтапно, в сравнении с переменными затратами их размер меняется незначительно.

**Прибыль** = Выручка - Затраты

**Выручка** = Количество транзакций * Средний чек 

**Затраты** = Постоянные + Переменные 

# Модель монетизации по принципу PNL (Profit and Losses):
    
Валовая прибыль = Выручка - Переменные затраты

Выручка = Количество транзакций * Средний чек

В PNL из схемы исключаются постоянные затраты. Такой расчет прибыли в разных подходах называется валовой прибылью, маржинальной или contribution margin.

Для подсчета **contribution margin** необходимы следующие метрики:

- UP, unit price – цена единицы продукта, сервиса или транзакции
- COGS, cost of goods sold — себестоимость на единицу продаж, или прямые расходы на единицу продукта (без учета САС). Очень часто также отдельно выделяется показатель first COGS, затраты на первую покупку — например, промокод как часть маркетинга.
- Количество единиц продукта на единицу клиента — если умножить цену продукта на эту метрику, то получится ARPPU
- САС — стоимость привлечения одного клиента
- Количество клиентов за период
- Постоянные затраты периода

В финансовой отчетности компаний за прибыль отвечает **EBITDA** (earnings before tax, depreciation and amortization) — прибыль до вычета расходов по кредитным ставкам, налогам и начисленной амортизации.
**EBITDA>0** фактически означает, что компания прибыльная и юнит-экономика сходится. Задача в том, чтобы увеличивать GMV и удерживать EBITDA.

# Финансовые отчеты показывают, окупается ли компания сейчас. Юнит-экономика показывает, что компания окупится в будущем.

Финансовый учет фиксирует фактическую прибыль и не объясняет, как и почему тот или иной бизнес зарабатывает. Помимо фактической прибыли, юнит-экономика показывает потенциальную прибыль и бизнес-процессы, которые позволят ее сформировать.

! Чтобы ваш бизнес был прибыльным, **маржинальная прибыль должна быть выше постоянных затрат.**

При запуске любая компания убыточна, и период окупаемости может оказаться достаточно долгим. **Как узнать, что стартап будет прибыльным?**

- Рассчитывать доход от потока клиентов, а не от каждого конкретного клиента
- Поточные данные разделять на единицы (юниты)
- Считать не только затраты на производство, но и на привлечение клиента

**Юнит-экономика помогает:**

- анализировать новые бизнес-идеи, продуктовые улучшения и стартапы: вы можете рассчитать, во сколько вам это обойдется и окупятся ли ваши затраты
- привлекать венчурные инвестиции: аргументировать, почему сейчас вы в убытке и как станете прибыльными. Начинающим предпринимателям авторы рекомендуют делать акцент на трёх составляющих: цена единицы продукции или доход на одного клиента, цена привлечения клиента, цена обслуживания клиента
- оценивать и управлять рекламными кампаниями (особенно, performance и digital). На заре промышленной революции достаточно было считать только выручку и затраты на производство продукта. В современном мире бизнес в большинстве случаев не выживет без маркетинга, поэтому нужно учитывать затраты на привлечение клиента

# Юнит-экономика (unit economics) — метод экономического моделирования, который помогает оценить ожидаемую прибыльность бизнес-идеи на основе данных.
Принципы юнит-экономики сжимают множество бизнес-процессов до относительно простых формул, строя данные путем расчета прибыльности единицы бизнеса (юнита). Юнит — это продажа или клиент.

# Основная идея юнит-экономики:
- прибыль — это выручка минус себестоимость и минус затраты на увеличение продаж/клиентов
- иными словами, если клиент часто возвращается и приносит денег больше, чем мы потратили на его привлечение, то мы все делаем правильно
- или на языке продуктовых метрик: Revenue = LTV - САС

LTV > CAC          :)

LTV < CAC          :(

Основной способ помочь бизнесу — сокращать САС и растить LTV

# Как прогнозировать ожидаемую прибыль:

# CM = UA x (ARPU — CPA), где:
- CM, Contribution Margin — объем денег, которые зарабатываем с потока пользователей с учетом наших расходов на себестоимость сделки.
- UA, User acquisition — число привлеченных пользователей в потоке.
- ARPU, Average Revenue Per User — сумма денег, которую приносит каждый пользователь в потоке.
- CoCA/CAC/CPA, Cost of Customer Acquisition — стоимость привлечения одного пользователя в поток.

Если **ARPU>CAC**, то компания зарабатывает. В противном случае мы несем убытки

Однако сами по себе эти метрики ничего не говорят о том, из-за чего как нам вырасти или из-за чего мы находимся в убытках. Чтобы принимать решения, нам необходимо связать эти метрики с метриками продукта, которые изменяются при
принятии решений.

**Оценка бизнес-моделей для freemium проектов:**

Прибыль(1) = цена(1) - себестоимость(1) - стоимость приобретения клиента - стоимость обслуживания клиента

# 3 составляющих успеха бизнес-модели: 
1) цена единицы продукции или доход на одного клиента 

2) стоимость приобретения клиента

3) стоимость обслуживания клиента 

Инвесторы оценивают жизнеспособность бизнес-модели как высокую, если удается достаточно просто объяснить как формируется прибыль, основываясь на 3-х вышеупомянутых тезисах.  

**Связь бизнес-метрик с метриками продукта**

Путь хорошей компании — от убытка к точке безубыточности. Как к ней прийти?
Давайте посмотрим на метрику ARPU. Это доход, который приносит каждый пользователь продукта — неважно, платит он или нет. Мы будем говорить об интернет-бизнесе, то есть сайте, и пользователем будем называть посетителя сайта.

Итак, **ARPU** — доход посетителя за некоторое время. 

Одно из определений LTV рассматривает связь ARPU и LTV:

**LTV = ARPU (LT)**, пожизненная ценность клиента равна сумме дохода от него за определенный срок

**ARPU = Revenue / UA**  : доход от 1 пользователя равен отношению выручки к числу пользователей в потоке

Еще один способ рассчитать значение ARPU — через ARPPU, доход от платящего пользователя. Связаны эти величины так же, как и посетители с платящими посетителями — через конверсию С1

**ARPU = ARPPU x C1**, где C1 — конверсия в первую покупку. Важно понимать, что нас интересует конверсия в первую покупку, так как она разделяет посетителей на простых и тех, кто хотя бы раз платил.

**ARPPU = (AOV - COGS) x APC - 1sCOGS, где:**
- AOV, Average Order Value — средний чек за выбранный промежуток времени 
- APC, Average Purchases of Customer — среднее число продаж на одного клиента. Это число включает в себя повторные сделки, которые совершают наши клиенты, и чем выше это число — тем лучше. Это означает, что клиенты готовы платить снова и снова
- 1sCOGS, first Cost of Goods Sold — особенные затраты на первую сделку, причем такие, которые не включены в COGS. Например, промокод на первую покупку, издержки на подключение клиента к сервису, тестовые периоды и тп.

# CM = UA x (ARPPU x C1 - CPA),
# CM = UA x (((AOV - COGS) x APC — 1sCOGS) x C1 — CAC)

# Связь бизнес-метрик с маркетинговыми метриками

**ROI (Return on Investment) или ROMI (Return on marketing investment)** — метрика, которая заменяет собой всю юнит-экономику. Она показывает, какой возврат вы получите при вложении денег в определенный канал дистрибуции.

**ROI = (LTV — CAC) / CAC**

# Оценка юнит экономики проекта (пример)

Вы запустили проект и собрали данные по проекту за 3 месяца:
- вы привлекаете на сайт 10000 посетителей в месяц
- цена 1 посетителя — 1 рубль
- конверсия посещения сайта в первую покупку — 0,5%
- средний чек 1 покупки — 1000 рублей
- средняя себестоимость корзины — 800 рублей
- Retention rate из месяца в месяц — 50%

❓ Нужно оценить юнит-экономику проекта. Сколько вы зарабатываете на одном клиенте?

Если считать только постоянные затраты, то на одном пользователе мы зарабатываем 200Р (1000-800), однако в терминах юнит-экономики нам нужно учесть затраты на привлечение. Чего нам стоит 1 покупка (1sCOGS)?

CPA = Затраты на привлечение / Поток пользователей х Конверсия в 1 покупку
= 10000 / ( 10000 х 0,005 ) = 200
Получается, что затраты на привлечение клиента равны нашему заработку при вычете себестоимости корзины. То есть после первой покупки выходим в ноль. Но как выглядит картина через несколько месяцев, если клиент возвращается?

Retention rate из месяца в месяц = 50%.

LTV: 200 + 0, 5 ∗ 200 + 0, 5^2

Revenue = LTV - CAC, то есть 400-200 = 200 рублей мы зарабатываем на каждом клиенте.

ROI = (LTV — CAC) / CAC , то есть 200/200=1 или 100%.

# Важно запомнить:
- Прибыль = CAC - LTV
- Каждая метрика содержит в себе как бизнесовый, так и продуктовый контекст. Поэтому мы можем смотреть на метрику под любым из этих углов, формулировать бизнес-метрики через продуктовые и наоборот. Если вы смотрите на несколько метрик, то вы всегда должны видеть гипотезу
- про продукт, бизнес или про их взаимосвязь. Анализируя метрики в разных контекстах, мы можем ответить на ключевые вопросы, например: какие свойства продукта дают наблюдаемый денежный результат? Что в продукте необходимо улучшить и чего нам это будет стоить? В чем основная задача и потребность пользователя и сколько мы заработаем на ее решении?
- Высокоуровневые метрики (САС, LTV) раскладываются на составляющие метрики — рычаги влияния на бизнес-процесс

- сегментируйте пользователей при расчете юнит-экономики
- юнит-экономику надо считать отдельно для разных каналов привлечения/ разных платформ/ разных регионов и тд

# Когортный анализ

- изучайте поведение и эффективность групп пользователей, имеющих общие атрибуты
- когорта - это группа пользователей, имеющих общее свойство, которое определеяется с помощью параметра Google Analytics. Например, все пользователи с одинаковой датой первого посещения относятся к одной когорте. 

**Когортный анализ помогает:**
- изучать отдельные когорты, чтобы оценить отдачу от краткосрочных маркетинговых мероприятий, таких как однодневные почтовые кампании
- отслеживать изменения в поведении и эффективности отдельных групп пользователей по дням, неделям, месяцам
- собирать пользователей в группы с учетом общих свойств, таких как дата первого посещения, а затем изучать поведение этих групп в соответствии с такими показателями как коэффициент удержания клиентов или доход.
