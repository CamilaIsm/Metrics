# Метрики

**Лекция 4. Маркетинговая аналитика**

**Что такое атрибуция пользователей**

Юнит-экономика может сходиться для одних каналов привлечения и не сходиться для других. Поэтому ее надо считать отдельно для разных каналов привлечения, для разных платформ, для разных регионов и т.д.

# 1.
Здесь возникает проблема качества данных. У используемых технологий есть свои особенности и ограничения, которые могут влиять на потерю событий. Например, при трекинге в WEB и приложениях существует так называемое окно атрибуции.

**Окно атрибуции** — это время, которое проходит с момента совершения действия пользователем и моментом, когда система отслеживания может нам об этом сообщить.

При трекинге в WEB пользователь мог заинтересоваться, но совершить целевое действие позже. И это влияет на процесс отслеживания эффективности — например, в моменте завершения потраченных денег по рекламной кампании необходимо подождать еще около двух недель, чтобы подгрузились все целевые действия.

При трекинге в App событие об установке приложения приходит на сервер не после собственно установки, а после того, как пользователь запустил приложение. В связи с этим установки мы можем посчитать по кварталу, а для байеров мы будем учитывать окно атрибуции в Х дней.

# 2. 
Если мы будем оценивать каждый канал по отдельности, то мы замножим пользователей и наша статистика будет необъективна.

Здесь возникает задача атрибуции пользователей. Нам нужно понять, какие
именно взаимодействия (показы, клики) и в каких каналах (соц.сети, сайт) привели
к продаже.

Атрибуция — процесс оценки влияния каждого взаимодействия рекламного инструмента с вашим клиентом в итоговую эффективность.

Основная модель атрибуции, используемая в том числе в рекламных кабинетах Google и Яндекс — модель последней конверсии (клика). То есть из всего пути клиента мы оцениваем последний этап (last click).

Проблема такой модели в том, что мы теряем множество данных о каналах, где пользователь взаимодействовал с нами, и получаем необъективную картину.
# Как атрибутировать пользователей

Каждая система распределяет эффективность только по тем  шагам , которые попадают к ней в систему.

Для того, чтобы определять эффективность отдельного канала по взаимодействиям, мы можем сопоставлять похожих по поведению пользователей.
Например, в Google есть два инструмента: в одном учитываются взаимодействия только с продуктами Google, а в другом можно интегрировать и учитывать внешние инструменты тоже

Используя эти инструменты, мы можем сопоставлять конверсии похожих путей пользователя (с разницей в один шаг) и оценить, насколько повлиял тот или иной канал

Тем не менее, такая модель остается необъективной: Youtube совместно с именно этими каналами дает конверсию 1.5%, однако не факт, что он даст 0.5% конверсии, если мы уберем или заменим один шаг в цепочке.
Иными словами, если мы попробуем оценить по такой концепции каждый канал в отдельности, то сумма вкладов конверсий будет больше, чем наблюдаемая нами конверсия.

Путь пользователя в некотором смысле является системой с наблюдаемым результатом — мы не можем наверняка оценить вклад каждого из ее элементов.

# Метрики в разрезе каналов привлечения

- Direct (Прямой/Type-in) — прямые заходы на сайт, в т.ч. закладки
- Organic Search (Органический поиск/SEO) — переходы из поисковых систем
- Paid Search (Платный поиск/Контекстная реклама) — Google Ads, Yandex Direct
- Display — баннерная и видео реклама
- Social (Соц сети) — переходы из социальных сетей
- Email + Push — электронная рассылка
- CPA/Partnerships (Партнерская программа) — реклама с оплатой за результат
- Referral (Реферальный трафик) — ссылки на ваш сайт с других сайтов
- Other (другое) — трафик, не попавший ни в один из каналов выше

# Важно запомнить

Для качественного маркетинга нужно:
- Хорошо работающие технические решения (качество данных) 
- Максимально четкое понимание, чего вы хотите добиться от вашей рекламы (целевые метрики)
- Стремиться отслеживать как можно больше этапов на пути к конверсии, вплоть до одобрения/выкупа заказа (атрибуция пользователей и оценка эффективности)
